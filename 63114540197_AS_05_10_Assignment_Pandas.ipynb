2. ให้แสดงสรุปค่าทางสถิติของข้อมูลที่ได้จากข้อ 1.

3. ให้แสดงกราฟขนาด (10,10) แบบกราฟเส้น (line chart) สีเขียว โดยระบุให้แกน x คือ area แกน y คือ price และระบุชื่อแกน y เป็น average price

```python
import pandas as pd
```

```python
df = pd.read_csv("baan4000.csv")
df_pvt = pd.pivot_table(df,values=["price","area"],index="province",aggfunc= "mean")
df2 = df_pvt.sort_values( "area", ascending=False)
my = df2[(df2[ "area"]>=60)&(df2[ "area"]<=100)]
my
```

```python
#2
df2 = pd.pivot_table(df,values=["area","price"],index="province",aggfunc={"sum" , "mean"})
df2
```

```python
#3
my2 = my.plot(kind='line', x='area', y='price', color='green',ylabel="average price",figsize=(10,10))
my2
```
